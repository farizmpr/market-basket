```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df_rules = pd.read_csv('rules.csv')
df_rules
```

```python
df_order = pd.read_csv('association_order.csv')
df_order
```

```python
def addaz(x):
    ret = "A"+str(x)+"Z"
    return(ret)
```

```python
df_order2 = df_order
df_order2['orderid'] = df_order2['orderid'].apply(addaz)
df_order2['itemid'] = df_order2['itemid'].apply(addaz)
df_order2.head()
```

```python
df_combined = df_order2.groupby(['orderid']).itemid.agg([('itemid', ', '.join)])
df_combined.reset_index(inplace=True)
df_combined.head(5)
```

```python
def hgsatusatu(item1,item2):
    df_count_s = df_combined[df_combined['itemid'].str.contains(addaz(item1))]
#     print("hanya 1 = "+str(df_count_s.shape[0]))
    df_count_c = df_combined[(df_combined['itemid'].str.contains(addaz(item1))) & (df_combined['itemid'].str.contains(addaz(item2)))]
    print("gabungan 1-2 = "+str(df_count_c.shape[0]))
    ret = int(df_count_c.shape[0]/df_count_s.shape[0]*1000)
    return(ret)
    
def hgsatudua(item1,item2,item3):
    df_count_s = df_combined[df_combined['itemid'].str.contains(addaz(item1))]
#     print("hanya 1 = "+str(df_count_s.shape[0]))
    df_count_c = df_combined[(df_combined['itemid'].str.contains(addaz(item1))) & (df_combined['itemid'].str.contains(addaz(item2))) & (df_combined['itemid'].str.contains(addaz(item3)))]
    print("gabungan 1-2-3 = "+str(df_count_c.shape[0]))
    ret = int(df_count_c.shape[0]/df_count_s.shape[0]*1000)
    return(ret)

def hgduasatu(item1,item2,item3):
    df_count_s = df_combined[df_combined['itemid'].str.contains(addaz(item1)) & (df_combined['itemid'].str.contains(addaz(item2)))]
#     print("hanya 1-2 = "+str(df_count_s.shape[0]))
    df_count_c = df_combined[(df_combined['itemid'].str.contains(addaz(item1))) & (df_combined['itemid'].str.contains(addaz(item2))) & (df_combined['itemid'].str.contains(addaz(item3)))]
    print("gabungan 1-2-3 = "+str(df_count_c.shape[0]))
    ret = int(df_count_c.shape[0]/df_count_s.shape[0]*1000)
    return(ret)
```

```python
def hitung(inp):
    splitted = inp.split('>')
    if('&' in splitted[0]):
        split2 = splitted[0].split('&')
        item1 = split2[0]
        item2 = split2[1]
        item3 = splitted[1]
        return(hgduasatu(item1,item2,item3))
    elif('&' in splitted[1]):
        split2 = splitted[1].split('&')
        item1 = splitted[0]
        item2 = split2[0]
        item3 = split2[1]
        return(hgsatudua(item1,item2,item3))
    else:
        item1 = splitted[0]
        item2 = splitted[1]
        return(hgsatusatu(item1,item2))
```

```python
dfbaru = df_rules
dfbaru['confidence']=df_rules['rule'].apply(hitung)
dfbaru
```

```python
dfbaru.to_csv()
```
